```python
    # number of members
    ens_NCEP = 15
    # size of the grid box
    # longitude for NCEP grid 1.5 deg 240 points
    dx_NCEP = 2 * np.pi * constant['R'] * np.cos(2 * np.pi * latitude_NCEP /
                                                 360) / 240 / 2 # weight by 2 to make it equivalent to our study with ERA-Interim grid
    # latitude for NCEP grid 1.5 deg 120 points
    dy_NCEP = np.pi * constant['R'] / 240 / 2 # weight by 2 to make it equivalent to our study with ERA-Interim grid
    
    # weight sea ice by area
    sic_NCEP_ens_area = np.zeros(sic_NCEP_ens.shape, dtype=float) #[ens, year, week, leadweek, lat, lon]
    for i in np.arange(len(latitude_NCEP[:])):
        # change the unit to km2
        sic_NCEP_ens_area[:,:,:,:,i,:] = sic_NCEP_ens[:,:,:,:,i,:]* dx_NCEP[i] * dy_NCEP / 1E+6 # unit km2
        
    # land sea mask correction for NCEP data using ERAI
    step_lead_NCEP = 6
    sea_ice_mask_3D = np.repeat(sea_ice_mask_barents[np.newaxis,:,:],step_lead_NCEP,0)
    sea_ice_mask_4D = np.repeat(sea_ice_mask_3D[np.newaxis,:,:,:],48,0)
    sea_ice_mask_5D = np.repeat(sea_ice_mask_4D[np.newaxis,:,:,:,:],3,0)
    sea_ice_mask_6D = np.repeat(sea_ice_mask_5D[np.newaxis,:,:,:,:,:],ens_NCEP,0)

    sic_NCEP_ens_area[sic_NCEP_ens_area<0] = 0
    sic_NCEP_ens_area = sic_NCEP_ens_area * sea_ice_mask_6D
    sic_NCEP_ens = sic_NCEP_ens  * sea_ice_mask_6D

    _, y, w, step_lead_NCEP, lat, lon = sic_NCEP_ens_area.shape
    sic_NCEP_ens_area_series = sic_NCEP_ens_area.reshape((ens_NCEP, y*w, step_lead_NCEP, lat, lon))
    sic_NCEP_ens_series = sic_NCEP_ens.reshape(ens_NCEP, y*w, step_lead_NCEP, lat, lon)
    
    print(sic_NCEP_ens_area_series.shape)
    sic_NCEP_ens_area_series_median = np.mean(sic_NCEP_ens_area_series,0)
    sic_NCEP_ens_series_median = np.mean(sic_NCEP_ens_series,0)
```

```python
    ##############################################################################################################
    ###########       Compute RMSE with given ensemble predictands and deterministic observation       ###########
    ##############################################################################################################
    step_lead = 6
    # data shape [ens, time, leadweek, lat, lon]
    # lead time dependent CRPS
    rmse = np.zeros((ens, step_lead), dtype=float)
    rmse_determin = np.zeros(step_lead, dtype=float)
    rmse_NCEP = np.zeros((ens_NCEP, step_lead), dtype=float)
    rmse_climatology = np.ones(step_lead, dtype=float)
    rmse_persist = np.zeros(step_lead, dtype=float)
    #crps_unit = np.zeros((test_year*12*4,step_lead,len(lat_cut),len(lon_cut)), dtype=float)
    for i in range(step_lead):
        for j in range(ens):
            if i == 0:
                rmse_inter, _ = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4+i:,:,:], lead_pred_sic_denorm[j,:,i,:,:])
                rmse[j,i] = np.mean(rmse_inter)
            else:
                rmse_inter, _  = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4+i:,:,:], lead_pred_sic_denorm[j,:-i,i,:,:])
                rmse[j,i] = np.mean(rmse_inter)
            
    for i in range(step_lead):
        for j in range(ens_NCEP):
            if i == 0:
                rmse_inter, _ = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4+i:,:,:], sic_NCEP_ens_series[j,:,i,:,:])
                rmse_NCEP[j,i] = np.mean(rmse_inter)
            else:
                rmse_inter, _ = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4+i:,:,:], sic_NCEP_ens_series[j,:-i,i,:,:])
                rmse_NCEP[j,i] = np.mean(rmse_inter)
    
    
    # take the mean rmse over ensemble
    rmse_mean = np.mean(rmse,0)
    rmse_NCEP_mean = np.mean(rmse_NCEP,0)
    
    # determinstic run
    for i in range(step_lead):
        if i == 0:
            rmse_determin_inter, _ = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4+i:,:,:],
                                                       lead_pred_sic_determin_denorm[:,i,:,:])
            rmse_determin[i] = np.mean(rmse_determin_inter)
        else:
            rmse_determin_inter, _ = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4+i:,:,:],
                                                       lead_pred_sic_determin_denorm[:-i,i,:,:])
            rmse_determin[i] = np.mean(rmse_determin_inter)            
    
    # rmse persistence
    for i in range(step_lead):
        rmse_persist_unit, _ = dlacs.metric.RMSE(sic_exp_denorm[-test_year*12*4-1:-1-i,:,:],
                                                 sic_exp_denorm[-test_year*12*4+i:,:,:])
        rmse_persist[i] = np.mean(rmse_persist_unit)
             
    # rmse climatology
    rmse_climatology_unit, _  = dlacs.metric.RMSE(climatology_seq_denorm,
                                                  sic_exp_denorm[-test_year*12*4:,:,:])
    rmse_climatology = rmse_climatology * np.mean(rmse_climatology_unit)
```
